```python
testData = JsonData("../20newsgroups").blogEntries[:100]
testData
```

```python
tdTransformations = {}
tdTransformations["tokenized"] = [nltk.tokenize.word_tokenize(blog) for blog in testData]
tdTransformations["tokenized"]
```

```python
stemmer = nltk.PorterStemmer()
tdTransformations["algoStemmed"] = [[stemmer.stem(word) for word in blog] for blog in tdTransformations["tokenized"]]
tdTransformations["algoStemmed"]
```

```python
import pickle

recreate = True
if recreate:
    stemmer = nltk.WordNetLemmatizer()
    tdTransformations["dictStemmed"] = [[stemmer.lemmatize(word) for word in blog] for blog in tdTransformations["tokenized"]]
    with open("py.cache") as file:
        pickle.dump(tdTransformations, file)
else:
    with open("py.cache") as file:
        tdTransformations  = pickle.load(tdTransformations, file)
    

tdTransformations["dictStemmed"]
```

```python
tdTransformations["taggedTokens"] = [nltk.pos_tag(blogTags) for blogTags in tdTransformations["tokenized"]]
tdTransformations["taggedTokens"]
```

```python
tdTransformations["tagFD"] = [nltk.FreqDist(tag for (word, tag) in blogTaggedTokens) for blogTaggedTokens in tdTransformations["taggedTokens"]]
tdTransformations["tagFD"]
```

```python
tdTransformations["tagFD"][1].plot()
tdTransformations["tagFD"][50].plot()
tdTransformations["tagFD"][99].plot()
```
